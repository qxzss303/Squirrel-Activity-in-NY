**Abstract**

**Background**

Squirrels are the most populated animal in Central Park. It is useful to have a good understanding of squirrel's behaviors and locations to help the city improve the park to better suit wildlife and promote a healthy ecosystem. The data of my research about squirrel's most common behaviors based on location comes from the 2018 Central Park Squirrel Census.

The research question I came up with is, "What are the most common behaviors of squirrels observed in different locations?" My hypothesis is that squirrels exhibit significantly different behaviors at different locations.
